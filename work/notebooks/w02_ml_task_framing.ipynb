# ML-03 — Frame Your Lane as an ML Task

This notebook frames the ML problem for the `fact_content_daily_performance` dataset.

## 1. My lane as an ML task (type)

**Task Type:** Ranking

The decision is to prioritize content pages for SEO optimization. SEO specialists use the output to decide which pages should be improved first. Ranking is appropriate because we want to order pages by priority.

```python
task_type='Ranking'
print('Task Type:',task_type)
print('Decision: Prioritize content pages for SEO optimization.')
print('Actor: SEO Specialist / Content Manager')
```

## 2. Target or proxy

The prediction target is **gsc_clicks**, an observed outcome from Google Search Console rather than a manually defined rule.

```python
target='gsc_clicks'
print('Prediction Target:',target)
print('Observed outcome from Google Search Console.')
```

## 3. Success metric

**Precision@K** is appropriate because the goal is to rank the highest-value pages near the top.

```python
metric='Precision@K'
print('Evaluation Metric:',metric)
```

## 4. The unit of analysis, as a real dataframe

One row represents the performance of one content page for one report date.

```python
from datasets import load_dataset
from itertools import islice
import pandas as pd

ds=load_dataset('FlyRank/internship-warehouse','fact_content_daily_performance',streaming=True,split='train')
df=pd.DataFrame(list(islice(ds,5)))
print(df.shape)
display(df)
```

## 5. Why ML beats a fixed rule here

Content performance depends on many interacting signals such as impressions, clicks, search position, engagement, and traffic sources. Machine learning can learn these relationships better than simple if-else rules and provides decision support.

```python
print('ML is suitable because content performance depends on many interacting signals that change over time.')
```
